```python
import librosa, librosa.display, librosa.util
import matplotlib.pyplot as plt
import numpy as np
```

```python
%cd D:\Data\扒谱小能手\AMT_dataset\spectrogram
```

```python
y, sr = librosa.load('../audio/StarS.wav')
C = np.abs(librosa.cqt(y, sr=sr, hop_length=64, n_bins = 84, bins_per_octave = 12))
```

```python
plt.subplot(2,1,1)
librosa.display.specshow(C, sr=sr, x_axis = 'time', y_axis = 'cqt_note')
plt.set_cmap('hot')
plt.title('Spectrogram using amplitude')
```

```python
plt.subplot(2,1,2)
librosa.display.specshow(librosa.amplitude_to_db(C), sr=sr, x_axis = 'time', y_axis = 'cqt_note')
plt.set_cmap('hot')
plt.title('Spectrogram using dB')
```

- 直接使用amplitude可更明显的显示基频  
- 使用转换后的db数据可将泛音频率细节展示出来
